```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Ze_read = pd.read_csv('data.csv')
```

```python
def calculate_iqr_and_bounds(data, column):

    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    Low = Q1 - 1.5 * IQR
    Up = Q3 + 1.5 * IQR
    
    return {
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "Lower Bound": Low,
        "Upper Bound": Up
    }
```

```python
columns_to_analyze = ["RSS", "currentLocationX", "currentLocationY",
                       "centerFreq0", "floor", "channelWidth", "frequency"]
```

```python
def visualize_outliers(data, column):

    plt.figure(figsize=(12, 6))
    sns.boxplot(x=data[column])
    plt.title(f"Boxplot of {column}")
    plt.show()
    
    
```

```python
for column in columns_to_analyze:
    print(f"Analysis for column: {column}")
    results = calculate_iqr_and_bounds(Ze_read, column)
    for key, value in results.items():
        print(f"{key}: {value}")
    print("\n")

    visualize_outliers(Ze_read, column)
```
