### Cleaning

```python
df = df.dropna(axis=1, thresh=df.isnull().sum())
df
```

```python
# Number of samples
df.count()
```

```python
# Number of features
df.count(axis=1)
```

### Transformation

```python
# after normalization
ndf = df.copy()

for column in ndf.select_dtypes(include='number'):
        ndf[column] = (ndf[column] - ndf[column].min()) / (ndf[column].max() - ndf[column].min())

ndf
```
